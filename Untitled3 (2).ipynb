```python
import pygame
from pygame.locals import *

class Square(pygame.sprite.Sprite):
    def __init__(self):
        super(Square, self).__init__()
        self.surf = pygame.Surface((25, 25))
        self.surf.fill((0, 200, 255))
        self.rect = self.surf.get_rect()

pygame.init()
screen = pygame.display.set_mode((800, 600))

square1 = Square()
square2 = Square()
square3 = Square()
square4 = Square()

screen.fill((0, 0, 0))
screen.blit(square1.surf, (40, 40))
screen.blit(square2.surf, (40, 530))
screen.blit(square3.surf, (730, 40))
screen.blit(square4.surf, (730, 530))
pygame.display.flip()

pygame.image.save(screen, "activity1_squares.png")
pygame.quit()

from IPython.display import Image
Image("activity1_squares.png")
```

```python
import pygame, imageio

pygame.init()
win = pygame.display.set_mode((500, 500))
pygame.display.set_caption("Scrolling Text")

Font = pygame.font.SysFont('timesnewroman', 30)
white = (255, 255, 255)
yellow = (255, 255, 0)
green = (0, 255, 255)
orange = (255, 100, 0)

letter1 = Font.render("C", False, orange, yellow)
letter2 = Font.render("O", False, orange, green)
letter3 = Font.render("M", False, orange, yellow)
letter4 = Font.render("S", False, orange, green)
letter5 = Font.render("A", False, orange, yellow)
letter6 = Font.render("T", False, orange, green)
letter7 = Font.render("S", False, orange, yellow)

frames = []
i, c = 0, 1

for frame_num in range(80):
    if i >= 820:
        i = 0
        c += 1

    win.fill(white)

    if c % 6 == 0:
        win.blit(letter1, (662 - i, -162 + i)); win.blit(letter2, (639 - i, -139 + i))
        win.blit(letter3, (608 - i, -108 + i)); win.blit(letter4, (579 - i, -79 + i))
        win.blit(letter5, (552 - i, -52 + i)); win.blit(letter6, (529 - i, -29 + i))
        win.blit(letter7, (500 - i, 0 + i))
    if c % 6 == 5:
        win.blit(letter1, (-162 + i, -162 + i)); win.blit(letter2, (-135 + i, -135 + i))
        win.blit(letter3, (-110 + i, -110 + i)); win.blit(letter4, (-79 + i, -79 + i))
        win.blit(letter5, (-52 + i, -52 + i)); win.blit(letter6, (-27 + i, -27 + i))
        win.blit(letter7, (0 + i, 0 + i))
    if c % 6 == 4:
        win.blit(letter1, (480, -180 + i)); win.blit(letter2, (480, -150 + i))
        win.blit(letter3, (480, -120 + i)); win.blit(letter4, (480, -90 + i))
        win.blit(letter5, (480, -60 + i)); win.blit(letter6, (480, -30 + i))
        win.blit(letter7, (480, 0 + i))
    if c % 6 == 3:
        win.blit(letter1, (0, -180 + i)); win.blit(letter2, (0, -150 + i))
        win.blit(letter3, (0, -120 + i)); win.blit(letter4, (0, -90 + i))
        win.blit(letter5, (0, -60 + i)); win.blit(letter6, (0, -30 + i))
        win.blit(letter7, (0, 0 + i))
    if c % 6 == 1:
        win.blit(letter1, (-124 + i, 0)); win.blit(letter2, (-102 + i, 0))
        win.blit(letter3, (-82 + i, 0)); win.blit(letter4, (-58 + i, 0))
        win.blit(letter5, (-40 + i, 0)); win.blit(letter6, (-19 + i, 0))
        win.blit(letter7, (0 + i, 0))
    if c % 6 == 2:
        win.blit(letter1, (-124 + i, 470)); win.blit(letter2, (-102 + i, 470))
        win.blit(letter3, (-82 + i, 470)); win.blit(letter4, (-58 + i, 470))
        win.blit(letter5, (-40 + i, 470)); win.blit(letter6, (-19 + i, 470))
        win.blit(letter7, (0 + i, 470))

    pygame.display.update()

    frame = pygame.surfarray.array3d(win)
    frame = frame.transpose([1, 0, 2])
    frames.append(frame)

    i += 20

pygame.quit()
imageio.mimsave("activity2_scrolling_text.gif", frames, fps=12)

from IPython.display import Image as IPImage
IPImage("activity2_scrolling_text.gif")
```

```python
from IPython.display import HTML

HTML("""
<div style="display:flex; flex-direction:column; align-items:center; font-family:Segoe UI, sans-serif; background:#1e272e; padding:25px; border-radius:14px;">
  <h2 style="color:#2ecc71; margin-bottom:5px;">🐍 Snake Game — Lab 13 Graded Task</h2>
  <div style="display:flex; gap:30px; margin-bottom:10px; font-size:18px; color:#dcdde1;">
    <div>Score: <span id="score" style="color:#2ecc71; font-weight:bold;">0</span></div>
    <div>Speed: <span id="speedLabel" style="color:#f39c12; font-weight:bold;">Normal</span></div>
  </div>
  <canvas id="game" width="500" height="500" style="border-radius:12px; box-shadow:0 8px 24px rgba(0,0,0,0.5); background:linear-gradient(135deg,#1e272e,#2f3640);"></canvas>
  <p style="color:#7f8fa6; margin-top:10px;">Click inside the canvas, then use Arrow Keys to move | Press R to restart</p>
  <div id="gameOverBox" style="display:none; margin-top:10px; padding:15px 25px; background:#e74c3c; color:white; border-radius:10px; font-size:20px; text-align:center;">
    💀 Game Over! Final Score: <span id="finalScore"></span><br>
    <span style="font-size:14px;">Press R to restart</span>
  </div>
</div>

<script>
const canvas = document.getElementById("game");
const ctx = canvas.getContext("2d");
const box = 20;
const cols = canvas.width / box;
const rows = canvas.height / box;

let snake, direction, food, bonusFood, walls, score, speed, gameInterval, gameOver;

function randomCell() {
  return { x: Math.floor(Math.random()*cols)*box, y: Math.floor(Math.random()*rows)*box };
}

function initGame() {
  snake = [{x:9*box,y:10*box},{x:8*box,y:10*box},{x:7*box,y:10*box}];
  direction = "RIGHT";
  food = randomCell();
  bonusFood = null;
  walls = [];
  score = 0;
  speed = 150;
  gameOver = false;

  document.getElementById("score").innerText = score;
  document.getElementById("speedLabel").innerText = "Normal";
  document.getElementById("gameOverBox").style.display = "none";

  if (gameInterval) clearInterval(gameInterval);
  gameInterval = setInterval(gameLoop, speed);
}

function spawnWalls() {
  walls = [];
  for (let i = 0; i < 4; i++) walls.push(randomCell());
}

function drawCell(x, y, color) {
  ctx.fillStyle = color;
  ctx.beginPath();
  ctx.roundRect(x+1, y+1, box-2, box-2, 5);
  ctx.fill();
}

function draw() {
  ctx.fillStyle = "#2f3640";
  ctx.fillRect(0, 0, canvas.width, canvas.height);
  walls.forEach(w => drawCell(w.x, w.y, "#7f8c8d"));
  drawCell(food.x, food.y, "#e74c3c");
  if (bonusFood) drawCell(bonusFood.x, bonusFood.y, "#f1c40f");
  snake.forEach((s,i) => drawCell(s.x, s.y, i===0 ? "#2ecc71" : "#27ae60"));
}

function gameLoop() {
  if (gameOver) return;
  let head = {...snake[0]};
  if (direction === "LEFT") head.x -= box;
  if (direction === "RIGHT") head.x += box;
  if (direction === "UP") head.y -= box;
  if (direction === "DOWN") head.y += box;

  if (head.x < 0 || head.x >= canvas.width || head.y < 0 || head.y >= canvas.height) return endGame();
  if (snake.some(s => s.x===head.x && s.y===head.y)) return endGame();
  if (walls.some(w => w.x===head.x && w.y===head.y)) return endGame();

  snake.unshift(head);
  let ate = false;
  if (head.x === food.x && head.y === food.y) { score += 10; food = randomCell(); ate = true; }
  if (bonusFood && head.x === bonusFood.x && head.y === bonusFood.y) { score += 20; bonusFood = null; ate = true; }
  if (!ate) snake.pop();

  document.getElementById("score").innerText = score;

  if (score > 100 && speed === 150) {
    speed = 75;
    document.getElementById("speedLabel").innerText = "Fast x2";
    clearInterval(gameInterval);
    gameInterval = setInterval(gameLoop, speed);
  }

  if (score > 150 && !bonusFood && walls.length === 0) {
    bonusFood = randomCell();
    spawnWalls();
  }

  draw();
}

function endGame() {
  gameOver = true;
  clearInterval(gameInterval);
  document.getElementById("finalScore").innerText = score;
  document.getElementById("gameOverBox").style.display = "block";
}

document.addEventListener("keydown", (e) => {
  if (e.key === "r" || e.key === "R") { initGame(); return; }
  if (gameOver) return;
  if (e.key === "ArrowLeft" && direction !== "RIGHT") direction = "LEFT";
  else if (e.key === "ArrowRight" && direction !== "LEFT") direction = "RIGHT";
  else if (e.key === "ArrowUp" && direction !== "DOWN") direction = "UP";
  else if (e.key === "ArrowDown" && direction !== "UP") direction = "DOWN";
});

initGame();
</script>
""")
```
